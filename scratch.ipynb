```r
library(tidyverse)
library(zeallot)
library(stringr)
library(magrittr)
```

```r
source("R/label.R")
```

```r
df = read_csv("data/raw_data.csv") %>%
    mutate(occupation = str_to_upper(tran_occ))
```

```r
mappings = read_csv("data/occupation_mapping_patterns.csv")
df_mapped = df %>% 
    mutate(occupation = label_vec(occupation, mappings))
df_mapped %>% 
count(occupation) %>%
arrange(desc(n)) 
```
